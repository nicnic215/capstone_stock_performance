```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader as web
```

```python
df_stk = pd.read_csv("/Users/Nicole/Desktop/FTDS/Capstone/SYK.csv",index_col='Date',parse_dates=True)
df_stk.head()
```

```python
df_rf = pd.read_csv("/Users/Nicole/Desktop/FTDS/Capstone/RiskFree/RiskFreeRate.csv", index_col='Date',parse_dates=True)
df_rf.head()
```

```python
df_stk['Returns'] = df_stk['Adj Close'].resample('BM').ffill().pct_change()
df_stk = df_stk.dropna()
#df_stk.head()
df_stk
```

```python
df_stk['Log_Ret'] = np.log(df_stk['Adj Close'] / df_stk['Adj Close'].shift(1))
```

```python
df_stk['Volatility'] = df_stk['Log_Ret'].rolling(window=4).std() * np.sqrt(12)
df_stk.tail(15)
```

```python
df_stk.head(20)
```

```python
df_stk = pd.concat([df_stk, df_rf], axis=1, sort=False)
```

```python
df_stk.head(20)
```

# Sharpe Ratio 
def sharpe(returns, rf, days=12):
    volatility = returns.std() * np.sqrt(days) 
    sharpe_ratio = (returns.mean() - rf) / volatility
    return sharpe_ratio

```python
# Sharpe Ratio 
def sharpe(returns, rf):
    volatility = df_stk['Volatility']
    sharpe_ratio = (returns.mean() - rf) / volatility
    return sharpe_ratio
```

```python
data = sharpe(df_stk['Returns'], df_stk['Risk Free'])
data
```

```python
data_new = pd.DataFrame(data)
data_new
```
